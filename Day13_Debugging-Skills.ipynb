### Generations - We used it as a reference
#### The Silent Generation: Born 1928-1945 (76-93 years old)
#### Baby Boomers: Born 1946-1964 (57-75 years old)
#### Generation X: Born 1965-1980 (41-56 years old)
#### Millennials: Born 1981-1996 (25-40 years old)
#### Generation Z: Born 1997-2012 (9-24 years old)
#### Generation Alpha: Born 2010-2025 (0-11 years old)

```python
# Play Computer
year = int(input("What's your year of birth?\n"))
if year > 1980 and year < 1997:# Be careful using comparison operators
  print("You are a millenial.")
elif year > 1996:
  print("You are a Gen Z.")
else:
    print("You are a silent generation")
```

```python
# Fix the Errors
age = int(input("How old are you?\n"))# We have to specify that age is an integer not a string which is default input in python
if age >= 18:
    print(f"You can drive at age {age}.")# f-string uses keyword "f" at the starting of a string # Indentation is compulsory for conditional blocks
else:
    print(f"You can not drive at age {age}.")
```

```python
#Print is Your Friend
pages = 0
word_per_page = 0
pages = int(input("Number of pages: "))
# word_per_page == int(input("Number of words per page: "))# "==" is a comparison operator not a assignment operator so we have to use "="
# print(word_per_page)# word_per_page doesn't changes its value as it has not been changed by any assignment operator
word_per_page = int(input("Number of words per page: "))
print(f"There are {pages} number of pages in the book and there are {word_per_page} words in each page")# Now we will get each value as we have assigned all of them above
total_words = pages * word_per_page
print(total_words)
```
